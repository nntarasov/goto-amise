```python
def get_class(x):
    if isinstance(x, float) and isnan(x):
        return 'undefined'
    for cur_class in classes:
        for cur_subclass in cur_class[1:]:
            if x.lower().find(cur_subclass) != -1:
                return cur_class[0]
    return 'undefined'
```

```python
def get_factors(data):
    result = defaultdict(dict)
    for user_id, actions in data.groupby('user_id'):
        in_acts = actions[actions['action_type'] == 'in']
        out_acts = actions[actions['action_type'] == 'out']
        cash_acts = in_acts[in_acts['category'].map(lambda x: x in ['QIWI', 'Терминалы и отделения партнеров QIWI'])]
        result[user_id]['cash_acts'] = len(cash_acts)
        max_time_str = actions['date_month'].max()
        result[user_id]['max_time'] = (int(max_time_str[:4]) - 2014) * 12 + int(max_time_str[5:])
        min_time_str = actions['date_month'].min()
        result[user_id]['min_time'] = (int(min_time_str[:4]) - 2014) * 12 + int(min_time_str[5:])
        result[user_id]['all_time'] = result[user_id]['max_time'] - result[user_id]['min_time'] + 1
        result[user_id]['office'] = out_acts[out_acts['category'] == 'Служебная']['payment'].sum() / result[user_id]['all_time']
        result[user_id]['credits'] = 0 if len(out_acts[out_acts['category'] == 'Погашение кредитов']) == 0 else 1
        result[user_id]['bookmaker_in'] = in_acts[in_acts['category'] == 'Букмекеры']['payment'].sum()
        result[user_id]['bookmaker_out'] = out_acts[out_acts['category'] == 'Букмекеры']['payment'].sum()
        result[user_id]['bookmaker_delta'] = result[user_id]['bookmaker_out'] - result[user_id]['bookmaker_in']
        result[user_id]['online_games_in'] = in_acts[in_acts['category'] == 'Онлайн игры']['payment'].sum()
        result[user_id]['online_games_out'] = out_acts[out_acts['category'] == 'Онлайн игры']['payment'].sum()
        result[user_id]['online_games_delta'] = result[user_id]['online_games_out'] - result[user_id]['online_games_in']
    return result
```

```python
factors = get_factors(all_data)
```

```python
import tensorflow as tf

sess = tf.Session()
W = tf.Variable(tf.zeros([len(features), 1]))
b = tf.Variable(tf.zeros([1]))
```

```python
X = tf.placeholder(tf.float32, [None, len(features)])
y = tf.placeholder(tf.float32, [None])
h = tf.nn.sigmoid(tf.matmul(X, W) + b)
loss = tf.reduce_mean(-y * tf.log(h + 1e-10) - (1-y) * tf.log(1-h + 1e-10))
train_step = tf.train.GradientDescentOptimizer(0.5).minimize(loss)
```

```python
sess.run(tf.global_variables_initializer())
```
